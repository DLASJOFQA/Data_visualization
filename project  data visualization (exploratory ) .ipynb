### Cleaning Data

```python
# Creating a copy of the original dataframe for cleaning.
df_clean = df_bike.copy()
# test
df_clean.info()
```

```python
# Choose columns that are important in the analysis only
df_clean = df_clean[['duration_sec', 'start_time', 'end_time', 'start_station_name', 'end_station_name', 
            'bike_id', 'user_type', 'member_birth_year', 'member_gender']]
# test
df_clean.head()
```

```python
# Converting several of the columns to the appropriate data type
df_clean.start_time = pd.to_datetime(df_clean.start_time)
df_clean.end_time = pd.to_datetime(df_clean.end_time)
df_clean.bike_id = df_clean.bike_id.astype(str)
df_clean.user_type = df_clean.user_type.astype('category')
df_clean.member_gender = df_clean.member_gender.astype('category')
# test
df_clean.dtypes
```

```python
# We can further enhance the dataset with more details about the time like month, day, hour, weekday
# this is start time
df_clean['start_time_month'] = df_clean['start_time'].dt.strftime('%B')
df_clean['start_time_weekday'] = df_clean['start_time'].dt.strftime('%a')
df_clean['start_time_hour'] = df_clean['start_time'].dt.strftime('%H')
# Extracting duration in minutes
df_clean['duration_min'] = df_clean['duration_sec']/60
df_clean['duration_min'] = df_clean['duration_min'].astype(int)
```

```python
# test
df_clean.info()
```

```python
# Extract the user's age in a new column year(2020 - member birth year)
df_clean['member_age'] = 2020 - df_clean['member_birth_year']
```

```python
# describe age
df_clean['member_age'].describe(percentiles=[0.99])
```

```python
# Remove outliers in the column for User birthdates (Dates prior to 1900).
df_clean = df_clean.query('member_age <= 65')
# test
print('member age maximum : ',df_clean['member_age'].max(),'\n','='*20)
# Convert the member_age and member_birth_year to integer format
df_clean.member_age = df_clean.member_age.astype(int)
df_clean.member_birth_year = df_clean.member_birth_year.astype(int)
# test
df_clean[['member_age', 'member_birth_year']].dtypes
```

## Univariate Exploration

> In this section, I am gonna investigate distributions of individual variables. This is indeed a preparation to look at the relationships between variables.

```python
# This function prints the percentage on the drawing
def print_presentation(counts , n_counts):
    """" It is a circular loop that walks along each row of data that performs the percentage calculation """
    for i in range (counts.shape[0]):
        count = counts[i]
        pct_string = '{:0.1f}%'.format(100*count/n_counts)
        plt.text(count+1, i, pct_string, va='center')
        
    from matplotlib import rcParams
    # Specify the figure size in inches, for both X, and Y axes
    rcParams['figure.figsize'] = 12,4
```

**Explore members gender and user type**

```python
plt.figure(figsize = [9, 6])
plt.subplot(1,2,1)
sorted_gender = df_clean['member_gender'].value_counts()
plt.pie(sorted_gender, labels=sorted_gender.index, startangle = 90, autopct = '%1.1f%%')

plt.subplot(1,2,2)
sorted_type = df_clean['user_type'].value_counts()
plt.pie(sorted_type, labels = sorted_type.index, startangle = 90, autopct = '%1.1f%%');
```

**Distribution of bike rides according to weekday,hour**

```python
# distribution weekday
plt.figure(figsize=[20,4])
plt.subplot(1,2,1)
count_day = df_clean.groupby('start_time_weekday').size()
weekday = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
color = sns.color_palette()[0]
sns.pointplot(data = df_clean, x = count_day.index, y = count_day , color = color , order = weekday);
plt.title("Distribution of bike rides according to day");

plt.subplot(1, 2, 2)
count_value = df_clean['start_time_weekday'].value_counts()
sns.countplot(data = df_clean, y='start_time_weekday', color=color, order= count_value.index);
plt.title('Propotion Start time weekday')
print_presentation(count_value , count_value.sum())
```

> **Observation : It is clear that there is a noticeable increase during the days of the week, with a decrease only on Saturday and Sunday, with a difference of about eight percent from the average participation of degrees during the days of the week, which indicates the use of the system during working days more than during holidays**

```python
#distribution to hour
plt.figure(figsize=[20,4])
plt.subplot(1,2,1)
count_hour = df_clean.groupby('start_time_hour').size()
sns.pointplot(data = df_clean,x =count_hour.index ,y=count_hour, color=color)
plt.title("Distribution of hourly bike rides")

plt.subplot(1,2,2)
sns.countplot(data=df_clean , x='start_time_hour', color=color);
plt.title("Distribution of hourly bike rides");
```

> **Observation :It is clear that there is a wave of rise during the early hours of the morning, specifically from seven to nine in the morning, then there is a decline with stability until another wave comes at the end of the work day, specifically from four in the evening until six in the evening. This also indicates that the system of participation Grades are in best condition during the first hours of work and with the end of work time**
